```python
df = df.loc[:, ~df.columns.str.startswith("V")]
df
```

```python
import numpy as np

# Step 1: Identify H-columns
h_columns = [col for col in df.columns if col.startswith("H")]

# Step 2: Make a copy of just the H data
h_data = df[h_columns].copy()

# Step 3: Find which rows have at least one non-NaN in H columns
has_data = h_data.notna().any(axis=1)

# Step 4: Interpolate only those rows, row-wise (axis=1)
interpolated = h_data[has_data].interpolate(axis=1, limit_direction='both')

# Step 5: Replace original rows with interpolated ones
df.loc[has_data, h_columns] = interpolated
df 
```

```python
import pandas as pd

# Load your correct file
df = pd.read_csv(r"C:\Users\Griselda\Desktop\Bootcamp Data Analytics\Sprint 10\Qualitat Aire - Mitjana.csv", sep=';', decimal=',')

# Select only H01-H24 columns
h_columns = [col for col in df.columns if col.startswith("H")]

# Group and mean ONLY for the hourly columns
mitjana_mes = df.groupby(
    ["ANY", "MES", "CODI_CONTAMINANT", "ESTACIO"]
)[h_columns].mean().reset_index()

# Save with correct format for Power BI
mitjana_mes.to_csv(
    r"C:\Users\Griselda\Desktop\Bootcamp Data Analytics\Sprint 10\Qualitat Aire - Mitjana.csv",
    sep=';',
    decimal=',',
    index=False
)

# Preview
print(mitjana_mes.head())

```
