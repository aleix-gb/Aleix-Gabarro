```python
import pandas as pd

transit = pd.read_csv(r"C:\Users\Griselda\Desktop\Bootcamp Data Analytics\Sprint 10\Evolució del trànsit a l'anell metropolità 2019-2023.csv")
transit.head()
```

```python
import pandas as pd

# Define grouping logic
weekday_group = ["Laborable", "Divendres"]
weekend_group = ["Dissabte", "Diumenge", "Festiu"]

# Drop rows with missing 'Tipus dia'
transit = transit.dropna(subset=["Tipus dia"])

# Create a new column for grouped day types
transit["Dia_agregat"] = transit["Tipus dia"].apply(
    lambda x: "Feiner" if x in weekday_group else "Cap de setmana/Festiu"
)

# Group by Year, Month, and the new Day Group
df_grouped = transit.groupby(["Any", "Mes", "Dia_agregat"], as_index=False)["Valor"].sum()

# (Optional) Pivot to get one row per month with both values
df_pivot = df_grouped.pivot(index=["Any", "Mes"], columns="Dia_agregat", values="Valor").reset_index()

# Show result
print(df_pivot.head())
```

```python
df_pivot = df_pivot[["Any", "Mes", "Feiner", "Cap de setmana/Festiu"]]
```

```python
df_pivot.rename(columns={
    "Feiner": "Feiner",
    "Cap de setmana/Festiu": "Festiu"
}, inplace=True)
df_pivot
```

```python
import pandas as pd

transit = pd.read_csv(r"C:\Users\Griselda\Desktop\Bootcamp Data Analytics\Sprint 10\Evolució del trànsit.csv", sep=';', decimal=',')
transit.head()
```

```python
# Ensure 'ANY' and 'MES' are integers
transit["Any"] = transit["Any"].astype(int)
transit["Mes"] = transit["Mes"].astype(int)

# Create yyyy-mm formatted column
transit["Data"] = transit["Any"].astype(str) + "-" + transit["Mes"].astype(str).str.zfill(2)
transit
```
